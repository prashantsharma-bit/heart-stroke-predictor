```python
import pandas as pd
import numpy as np

df=pd.read_csv('heart_disease.csv')  # replace with your file path
print(df)

```

```python
print(df.isnull().sum())
```

```python
df["education"]=df["education"].replace(np.nan,"unknown")
#print(df.isnull().sum())
```

```python
df["BPMeds"] = df["BPMeds"].fillna(df["BPMeds"].median())
#print(df.isnull().sum())
```

```python
df["totChol"] = df["totChol"].fillna(df["totChol"].median())
#print(df.isnull().sum())
#print(df["totChol"].median())
```

```python
df['cigsPerDay'] = df['cigsPerDay'].fillna(df['cigsPerDay'].median())
#print(df.isnull().sum())
```

```python
df["BMI"] = df["BMI"].fillna(df["BMI"].median())
#print(df.isnull().sum())
#print(df["BMI"].median())
```

```python
df["heartRate"] = df["heartRate"].fillna(df["heartRate"].median())
#print(df.isnull().sum())
print(df["heartRate"].median())
```

```python
df["glucose"] = df["glucose"].fillna(df["glucose"].median())
#print(df.isnull().sum())
#print(df["glucose"].median())
```

```python
print(df.isnull().sum())
```

```python
#after cleaning
print(df.notnull())
```
